```python
Y_train_list=train_data.iloc[:,1].tolist()
X_train_list=train_data.iloc[:,7].tolist()
```

```python
textFiles=np.asarray(X_train_list)
Y_train=np.asarray(Y_train_list)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
documents = []
i=0
for textFile in textFiles:
    documents.append((word_tokenize(textFile),Y_train[i]))
    i=i+1
documents[0:5]
```

```python
documents[0]
```

```python
from nltk.corpus import stopwords
import string
```

```python
stop=stopwords.words("english")
punctuation=list(string.punctuation)
stop=stop+punctuation
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(document), category) for document, category in documents]
```

```python
print(documents[0])
```

```python
X_train=np.array([" ".join(document) for document, category in documents])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer(max_features = 2000)
x_train_features = count_vec.fit_transform(X_train)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
X_test_list=test_data.iloc[:,6].tolist()
test_files_test=np.array(X_test_list)
```

```python
documentsTest = [clean_review(document) for document in documentsTest]
```

```python
x_test_features = count_vec.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_features, Y_train)
```
